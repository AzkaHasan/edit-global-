```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("/content/global_co2.csv")
dataset
```

```python
#df=dataset.fillna(value=dataset['Per Capita'].mean())
#df
```

```python
#dataset['Per Capita'] = dataset['Per Capita'].fillna(dataset['Per Capita'].expanding().mean())
#dataset
```

```python
dataset['Per Capita'] = dataset['Per Capita'].fillna('0')
dataset
```

```python
dataset['Per Capita'] = dataset['Per Capita'].astype(float)
dataset
```

```python
x= dataset.iloc[:,0:1].values

y= dataset.iloc[:,-1].values
print(x)
print(y)
```

Apply Train-Test-split method

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

1- Apply Linear Regression:

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train,color='red', marker='o', s=20,)
plt.plot(x_train,regressor.predict(x_train),color='blue')

plt.show()
```

```python
print(regressor.predict([[2011]]))
```

```python
print(regressor.predict([[2012]]))
```

```python
print(regressor.predict([[2013]]))
```

2-Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y_train)
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, lin_reg_2.predict(poly_reg.fit_transform(x_train)), color = 'blue')
plt.show()
```

```python
print(lin_reg_2.predict(poly_reg.fit_transform([[2011]])))
```

```python
print(lin_reg_2.predict(poly_reg.fit_transform([[2012]])))
```

```python
print(lin_reg_2.predict(poly_reg.fit_transform([[2013]])))
```

4- Random Forest tree

```python
from sklearn.ensemble import RandomForestRegressor
regressor2 = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor2.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor2.predict(x_train), color = 'blue')
plt.show()
```
